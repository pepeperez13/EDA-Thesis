```python
import pandas as pd
from IPython.display import display
```

```python
# Read all data from 2018
demands_df = pd.read_csv(r"D:\Master\Thesis\Code\L-Town\2018_SCADA_Demands.csv", sep = ";", decimal = ",") # Junctions demand Liters per hour (L/h) (nodeID)
flows_df = pd.read_csv(r"D:\Master\Thesis\Code\L-Town\2018_SCADA_Flows.csv", sep = ";", decimal = ",") # Links flow Cubic meters per hour(m^3/h) (linkID)
levels_df = pd.read_csv(r"D:\Master\Thesis\Code\L-Town\2018_SCADA_Levels.csv", sep = ";", decimal = ",") # Tanks level of water Meters (m). (tankNodeID)
pressures_df = pd.read_csv(r"D:\Master\Thesis\Code\L-Town\2018_SCADA_Pressures.csv", sep = ";", decimal = ",") # Junctions pressure Meters (m). (nodeID)
```

```python
display(demands_df, flows_df, levels_df, pressures_df)
```

```python
# Demands data in long format and with unique identifier
demands_long = demands_df.melt(id_vars=["Timestamp"], var_name="sensor_id", value_name="measurement")
demands_long["sensor_id"] = (demands_long["sensor_id"] + "_demand")
demands_long["sensor_type"] = "demand"
demands_long
```

```python
# Flows data in long format with unique identifier
flows_long = flows_df.melt(id_vars=["Timestamp"], var_name="sensor_id", value_name="measurement")
flows_long["sensor_id"] = (flows_long["sensor_id"] + "_flow")
flows_long["sensor_type"] = "flow"
flows_long
```

```python
# Water level in T1 in long format with unique identifier
levels_long = levels_df.copy().rename(columns = {"T1": "measurement"})
levels_long["sensor_id"] = ("T1" + "_level")
levels_long["sensor_type"] = "level"
levels_long
```

```python
# Pressures in nodes in long format with unique identifier
pressures_long = pressures_df.melt(id_vars=["Timestamp"], var_name="sensor_id", value_name="measurement")
pressures_long["sensor_id"] = (demands_long["sensor_id"] + "_pressure")
pressures_long["sensor_type"] = "pressure"
pressures_long
```

```python
ltown_df = pd.concat([demands_long, flows_long, levels_long, pressures_long], ignore_index=True)
ltown_df['unique_id'] = ltown_df['sensor_id'] + '_' + ltown_df["Timestamp"].astype(str)
ltown_df
```
